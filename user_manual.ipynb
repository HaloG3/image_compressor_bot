**User Manual: AI Image Compressor**

---

## **Introduction**
THis * Image Compressor Bot* is a Windows application that allows users to compress images efficiently using ML-based techniques while preserving their quality. The application works offline and supports multiple image formats, including JPEG, PNG, GIF, and BMP. Users can select multiple images, adjust compression quality using a slider, preview images before saving, and process images in batches.

## **System Requirements**
- Windows 10 or later
- Python 3.8+
- Dependencies: PyQt5, PIL (Pillow),  OpenCV
- GPU (optional for better performance)

## **Installation Instructions**
1. Install Python 3.8 or later.
2. Install dependencies using:
   ```sh
   pip install PyQt5 Pillow tensorflow opencv-python
   ```
3. Run the application:
   ```sh
   python main.py
   ```

## **Features Overview**
- **Batch Image Processing**: Compress multiple images at once.
- **Adjustable Compression Levels**: Slider to set compression percentage.
- **Image Preview**: View selected images before saving.
- **Preview of the image** : get preview of the image
- **Save Options**: Save all at one location ,selected,  or single images at different location .
- **Windows Explorer Integration**: Easy access to compressed images.

## **How to Use**
0. **Download Link** https://archive.org/download/image-compressor-bot_202503 
1. **Launch the Application**
   - Double-click the `.exe` file (if compiled) or run `python main.py`.
2. **Add Images**
   - Click **Add Images** or drag-and-drop images into the input box.
3. **Adjust Compression**
   - Use the slider to set the compression percentage (default: 70%).
4. **Compress Images**
   - Click **Compress Images** to start the process.
5. **Preview & Save**
   - Click on any compressed image to preview it.
   - Click **Save Selected Image** to save your file.
